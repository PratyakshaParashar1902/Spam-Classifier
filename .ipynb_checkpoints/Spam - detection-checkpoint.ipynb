## 3. Data Preprocessing
- Lower case
- Tokenization
- Removing special characters
- Removing stop words and punctuation
- Stemming
- def transform_

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
string.punctuation
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# eg
ps.stem('Loving')
```

```python
# Now lets write a function that will perform the above written transformation on our data.
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
#   Now text is in the form of list

    y = []
    for i in text:
#         isalnum is that i is alpha-numeric.
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
        
#     Returning as a string
    return " ".join(y)
```

```python
# Lets check transform_text
transform_text("I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k?")
```

```python
df['transformed_text'] = df['text'].apply(transform_text)
```

## 4. Model Building

#### It is seen that Naive Base algorith works well on the textual data so our first model will be on it. Algorithm take numeric 
#### input so we have to convert out transformed_text into vector.

#### Ways to convert text into vector -
1. Bag of Words (BoW)
- The Bag of Words model represents text as a collection of word frequencies. Each word in the text is treated as a feature, and the frequency of each word in the text is counted.

2. TF-IDF (Term Frequency-Inverse Document Frequency)
- TF-IDF is a statistical measure used to evaluate how important a word is to a document in a collection of documents. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus.

3. Word Embeddings (Word2Vec, GloVe)
- Word embeddings are dense vector representations of words. Word2Vec and GloVe are popular techniques for generating word embeddings. These models capture semantic relationships between words.

4. One-Hot Encoding
- One-hot encoding represents each word as a binary vector, where each dimension corresponds to a word in the vocabulary. The vector is 1 at the index corresponding to the word and 0 elsewhere.

5. N-grams
- N-grams are contiguous sequences of n items from a given sample of text. They capture local word order information. Both CountVectorizer and TfidfVectorizer support n-grams.
